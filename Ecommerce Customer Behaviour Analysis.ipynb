#### Keep the transactions seperate like,morning,afternoon,evening,night

```python
def order_time(x):
    if x in [6,7,8,9,10,11,12]:
        return 'Morning'
    if x in [13,14,15,16,17]:
        return 'Afternoon'
    if x in [18,19,20,21,22]:
        return 'Evening'
    if x in [23,24,0,1,2,3,4,5]:
        return 'Night'
    return x
```

```python
df['order_time_list']=df['order_hour_of_day'].apply(order_time)
```

```python
df.head()
```

```python
res2 = df.groupby('order_time_list')['user_id'].agg(['count'])
```

```python
res2
```

```python
res2.plot(
    kind = 'bar',
    title = 'Time of the day when, the order was made'
)
```

#### The number of orders that users have made

```python
df.groupby('order_number')['user_id'].agg(['count'])
```

```python
def order_number_group(x):
    if x < 20:
        x = '1-20 orders'
    elif x > 20 and x <= 40:
        x = '21-40 orders'
    elif x > 40 and x <= 60:
        x = '41-60 orders'
    elif x > 60 and x <= 80:
        x = '61-80 orders'
    else:
        x = '81-100 orders'
    return x
```

```python
df['order_number_group'] = df['order_number'].apply(order_number_group)
```

```python
df
```

```python
res4 = df.groupby('order_number_group')['user_id'].agg(['count']).sort_values(by = 'count' ,ascending = False)
```

```python
res4
```

```python
res4.plot(
    kind  = 'pie',
    autopct = '%1.3f%%',
    subplots = True,
    title = 'Number of orders',
    figsize = (8,8)
)
```
